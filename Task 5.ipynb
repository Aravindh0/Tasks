1. Import the necessary libraries (Matplotlib and Seaborn) and load the data from the provided CSV file into a
Pandas DataFrame.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('D:\data science/padas_question.csv')
df
```

2. Create a scatter plot using Matplotlib to visualize the relationship between height and weight in the dataset. Label
the axes and give the plot a title.

```python
x=df['Height (cm)']
y=df['Weight (kg)']
plt.figure(figsize=(10,6))
plt.scatter(x,y)
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Relationship between height and weight')
plt.show()
```

3. Create a histogram using Matplotlib to visualize the distribution of ages in the dataset. Label the axes and give
the plot a title.

```python
x=df['Age']
plt.figure(figsize=(10,6))
plt.hist(x)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of ages')
plt.show()
```

4. Create a box plot using Seaborn to visualize the distribution of heights in the dataset by gender. Label the axes
and give the plot a title.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df['Height (cm)'],y=df['Gender'])
plt.xlabel('Height (cm)')
plt.ylabel('Gender')
plt.title('Distribution of height by gender')
plt.show()
```

5. Create a bar chart using Seaborn to visualize the average weight of each gender in the dataset. Label the axes
and give the plot a title.

```python
plt.figure(figsize=(10,6))
sns.barplot(x=df['Weight (kg)'],y=df['Gender'])
plt.xlabel('Weight (kg)')
plt.ylabel('Gender')
plt.title('Average weight of each gender')
plt.show()
```
